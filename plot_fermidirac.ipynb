```python
from scipy.special import expit
import matplotlib.pyplot as plt
import numpy as np
```

```python
distances = np.array([0,1,2,3,4,5])
mu = [0.01,0.05,0.1,0.5,0.9,1.5]
temp = [0.01,0.05,0.1,0.15,0.2,0.25,0.3,0.4]
# mu 0.5, temp 0.1
# mu 0.3, temp 0.1
# mu 0.7, temp 0.1
# choose range temp (0.01,0.3)
# choose range mu (0.01,1)
for mu_i in mu:
    for temp_i in temp:
        
        exponents = (distances / max(distances) - mu_i) / temp_i
        probs = expit(-exponents)
        
        plt.figure()
        plt.plot(distances,probs)
        plt.title(f"mu{mu_i},temp{temp_i}")
        plt.show()
```
